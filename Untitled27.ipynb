```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Optional for sentiment
from textblob import TextBlob

# Display plots in notebook
%matplotlib inline

# Read data
df = pd.read_csv('USvideos.csv')
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
df.columns
```

```python
# Convert dates (optional)
df['trending_date'] = pd.to_datetime(df['trending_date'], errors='coerce')
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')
```

```python
df[['title', 'channel_title', 'views']].sort_values(by='views', ascending=False).head(10)
```

```python
top = df.sort_values(by='views', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='views', y='title', data=top)
plt.title('Top 10 Most Viewed Videos')
plt.xlabel('Views')
plt.ylabel('Video Title')
plt.show()
```

```python
def get_sentiment(text):
    return TextBlob(str(text)).sentiment.polarity

df['title_sentiment'] = df['title'].apply(get_sentiment)
sns.histplot(df['title_sentiment'], bins=30, kde=True)
plt.title("Sentiment Distribution of Video Titles")
```

```python
import pandas as pd

# Load each file
us_df = pd.read_csv('USvideos.csv')
in_df = pd.read_csv('INvideos.csv')
gb_df = pd.read_csv('GBvideos.csv')
ca_df = pd.read_csv('CAvideos.csv')

# Add region column
us_df['region'] = 'US'
in_df['region'] = 'IN'
gb_df['region'] = 'GB'
ca_df['region'] = 'CA'
```

```python
combined_df = pd.concat([us_df, in_df, gb_df, ca_df], ignore_index=True)
combined_df.to_csv('AllRegions_Trending.csv', index=False)
combined_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot to compare views by region
plt.figure(figsize=(10, 5))
sns.boxplot(data=combined_df, x='region', y='views')
plt.title('Views Distribution by Region')
plt.show()
```
